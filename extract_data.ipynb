```python
import pandas as pd
import numpy as np
import os
import glob
import shutil

os.getcwd()
path = '/Users/mona/Documents/all/Documents/programmin/jupyter/lipids/'
os.chdir(path)

df_1 = pd.read_excel('PI_1.xlsx', skiprows=4, sheet_name= "Relative")
df_2 = pd.read_excel('PI_2.xlsx', skiprows=4, sheet_name= "Relative")
```

```python
df_1.columns = df_1.iloc[0, :]
df_2.columns = df_2.iloc[0, :]
```

```python
df_1.drop(df_1.index[0], inplace=True)
df_1.drop(df_1.columns[[0,2,3]], axis=1, inplace=True)
df_2.drop(df_2.index[0], inplace=True)
df_2.drop(df_2.columns[[0,2,3]], axis=1, inplace=True)
```

```python
df= pd.merge(df_1, df_2, on="Lipid ID")
```

```python
tumor_list = []
normal_list = []
for k in df.columns:
    if k.endswith("T"):
        tumor_list.append(k)
    elif k.endswith("N"):
        normal_list.append(k)
        
```

```python
df_tumor = df.copy()
for j in normal_list:
    df_tumor.drop([j], axis=1, inplace=True)
```

```python
df_normal = df.copy()
for j in tumor_list:
    df_normal.drop([j], axis=1, inplace=True)
```

```python
df_normal["Mean"] = df_normal.loc[:, normal_list].mean(axis=1)
df_normal["STD"] = df_normal.loc[:, normal_list].std(axis=1)
df_tumor["Mean"] = df_tumor.loc[:, tumor_list].mean(axis=1)
df_tumor["STD"] = df_tumor.loc[:, tumor_list].std(axis=1)
```

```python
df_tumor.sort_values(by=['Mean'], ascending=False)
```

```python
df_normal.sort_values(by=['Mean'], ascending=False)
```

```python
df_normal_f = df_normal.sort_values(by=['Mean'], ascending=False).iloc[0:10, [0,-2,-1]]
df_tumor_f = df_tumor.sort_values(by=['Mean'], ascending=False).iloc[0:10, [0,-2,-1]]
```

```python
df_normal_f.reset_index(drop=True, inplace=True)
df_tumor_f.reset_index(drop=True, inplace=True)
```

```python
df_final = pd.concat([df_normal_f,df_tumor_f], keys= ["Normal", "Tumor"], axis=1)
                    
```

```python
df_final.to_excel("percentage_PI.xlsx",  sheet_name= "PI")
```
